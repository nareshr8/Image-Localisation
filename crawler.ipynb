## Crawl Image to pull background artworks

Here we are crawling Google, Bing to get images of various artwork. We will then superimpose our stamp on to it as a way to generate the stamped images. For this we are using a python library called **icrawler**. 

As a first step we are installing the icrawler library. I am commenting this out as a general practise so that it doesnt autorun each time I run the entire notebook.

Import the libraries.

```python
from icrawler.builtin import  (BingImageCrawler,GoogleImageCrawler)
import logging
```

The keyword to search for and crawl images.

```python
keyword = 'grocery item posters'
```

We will have images from each search engine in seperate folder and then consolidate them later together after verifying and discriminating the images if we have issues. We are downloading 1000 images at as its the maximum limit set by search engines on their API

```python
directory_google = './data/stamp-localisation/google-grocery'

# search_filters_google = dict(
#         date=((2016, 5, 13), None))
search_filters_google = dict()

google_crawler = GoogleImageCrawler(downloader_threads=4,storage={'root_dir': directory_google},log_level=logging.INFO)
google_crawler.crawl(keyword=keyword, filters=search_filters_google, max_num=1000)
```

Download from Bing

```python
search_filters_bing = dict()
directory_bing = './data/stamp-localisation/bing-grocery'

bing_crawler = BingImageCrawler(downloader_threads=4,storage={'root_dir': directory_bing},log_level=logging.INFO)
bing_crawler.crawl(keyword=keyword, filters=search_filters_bing, max_num=1000)
```
